# **Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.**

```python
import numpy as np
from scipy import stats

def perform_z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test and interprets the results.

    Args:
        sample_data (list or numpy.ndarray): The sample data.
        population_mean (float): The known population mean.
        population_std (float): The known population standard deviation.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): 'two-sided', 'less', or 'greater'. Defaults to 'two-sided'.

    Returns:
        tuple: (z_statistic, p_value, interpretation)
    """

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'less', or 'greater'.")

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. There is significant evidence that the sample mean is different from the population mean."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. There is not enough evidence to conclude that the sample mean is different from the population mean."

    return z_statistic, p_value, interpretation

sample_data = [102, 105, 98, 108, 100, 103, 99, 101, 106, 104]
population_mean = 100
population_std = 5
alpha = 0.05
alternative = 'two-sided'

z_stat, p_val, result_interpretation = perform_z_test(sample_data, population_mean, population_std, alpha, alternative)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

alternative = 'greater'
z_stat, p_val, result_interpretation = perform_z_test(sample_data, population_mean, population_std, alpha, alternative)

print(f"\nOne Sided Test (greater than)")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

alternative = 'less'
z_stat, p_val, result_interpretation = perform_z_test(sample_data, population_mean, population_std, alpha, alternative)

print(f"\nOne Sided Test (less than)")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)
```

# **Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_hypothesis_errors(null_mean, alt_mean, std_dev, sample_size, alpha, power=None):
    """
    Calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

    Args:
        null_mean (float): The mean under the null hypothesis.
        alt_mean (float): The mean under the alternative hypothesis.
        std_dev (float): The standard deviation of the population.
        sample_size (int): The sample size.
        alpha (float): The significance level (Type 1 error rate).
        power (float, optional): The power of the test (1 - Type 2 error rate). If None, it's calculated.
    """

    critical_value = null_mean + stats.norm.ppf(1 - alpha) * (std_dev / np.sqrt(sample_size))

    if power is None:
        beta = stats.norm.cdf(critical_value, loc=alt_mean, scale=std_dev / np.sqrt(sample_size))
        power = 1 - beta
    else:
        beta = 1 - power

    x = np.linspace(null_mean - 4 * (std_dev / np.sqrt(sample_size)),
                    alt_mean + 4 * (std_dev / np.sqrt(sample_size)), 1000)
    y_null = stats.norm.pdf(x, loc=null_mean, scale=std_dev / np.sqrt(sample_size))
    y_alt = stats.norm.pdf(x, loc=alt_mean, scale=std_dev / np.sqrt(sample_size))

    plt.plot(x, y_null, label='Null Hypothesis Distribution')
    plt.plot(x, y_alt, label='Alternative Hypothesis Distribution')

    plt.fill_between(x, y_null, where=(x >= critical_value), color='red', alpha=0.3, label=f'Type 1 Error (Î±) = {alpha:.3f}')

    plt.fill_between(x, y_alt, where=(x <= critical_value), color='orange', alpha=0.3, label=f'Type 2 Error (Î²) = {beta:.3f}')

    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.title('Type 1 and Type 2 Errors')
    plt.legend()
    plt.grid(True)
    plt.show()

null_mean = 100
alt_mean = 105
std_dev = 15
sample_size = 30
alpha = 0.05

visualize_hypothesis_errors(null_mean, alt_mean, std_dev, sample_size, alpha)
visualize_hypothesis_errors(null_mean, alt_mean, std_dev, sample_size, alpha, power = 0.8)
```

# **Q6. Write a Python program to perform an independent T-test and interpret the results.**

```python
import numpy as np
from scipy import stats

def independent_t_test(group1, group2, alpha=0.05, alternative='two-sided'):
    """
    Performs an independent two-sample t-test and interprets the results.

    Args:
        group1 (list or numpy.ndarray): The data for the first group.
        group2 (list or numpy.ndarray): The data for the second group.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): 'two-sided', 'less', or 'greater'. Defaults to 'two-sided'.

    Returns:
        tuple: (t_statistic, p_value, interpretation)
    """

    t_statistic, p_value = stats.ttest_ind(group1, group2, equal_var=False, alternative=alternative)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. There is significant evidence that the means of the two groups are different."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. There is not enough evidence to conclude that the means of the two groups are different."

    return t_statistic, p_value, interpretation

group1 = [25, 30, 28, 35, 32, 29]
group2 = [20, 26, 23, 28, 25, 22]

t_stat, p_val, result_interpretation = independent_t_test(group1, group2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

t_stat, p_val, result_interpretation = independent_t_test(group1, group2, alternative='greater')
print("\nOne sided greater than:")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

t_stat, p_val, result_interpretation = independent_t_test(group1, group2, alternative='less')
print("\nOne sided less than:")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)
```

# **Q8. Simulate data and perform both Z-test and T-test, then compare the results using Python.**

```python
import numpy as np
from scipy import stats

def simulate_and_compare_tests(sample_size, population_mean, population_std, alpha=0.05):
    """
    Simulates data, performs Z-test and T-test, and compares the results.

    Args:
        sample_size (int): The size of the simulated sample.
        population_mean (float): The known population mean.
        population_std (float): The known population standard deviation.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        dict: A dictionary containing the results of the Z-test and T-test.
    """

    simulated_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

    sample_mean = np.mean(simulated_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    t_statistic, p_value_t = stats.ttest_1samp(simulated_data, population_mean)

    results = {
        "z_test": {
            "z_statistic": z_statistic,
            "p_value": p_value_z,
            "decision": "Reject" if p_value_z < alpha else "Fail to reject",
        },
        "t_test": {
            "t_statistic": t_statistic,
            "p_value": p_value_t,
            "decision": "Reject" if p_value_t < alpha else "Fail to reject",
        },
    }

    return results

sample_size = 30
population_mean = 100
population_std = 15

results = simulate_and_compare_tests(sample_size, population_mean, population_std)

print("Z-test Results:")
print(f"  Z-statistic: {results['z_test']['z_statistic']:.4f}")
print(f"  P-value: {results['z_test']['p_value']:.4f}")
print(f"  Decision: {results['z_test']['decision']}")

print("\nT-test Results:")
print(f"  T-statistic: {results['t_test']['t_statistic']:.4f}")
print(f"  P-value: {results['t_test']['p_value']:.4f}")
print(f"  Decision: {results['t_test']['decision']}")

sample_size = 10
results = simulate_and_compare_tests(sample_size, population_mean, population_std)

print("\nSmall Sample Size Results:")
print("Z-test Results:")
print(f"  Z-statistic: {results['z_test']['z_statistic']:.4f}")
print(f"  P-value: {results['z_test']['p_value']:.4f}")
print(f"  Decision: {results['z_test']['decision']}")

print("\nT-test Results:")
print(f"  T-statistic: {results['t_test']['t_statistic']:.4f}")
print(f"  P-value: {results['t_test']['p_value']:.4f}")
print(f"  Decision: {results['t_test']['decision']}")
```

# **Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Args:
        data (list or numpy.ndarray): The sample data.
        confidence_level (float, optional): The desired confidence level (e.g., 0.95 for 95%). Defaults to 0.95.

    Returns:
        tuple: (lower_bound, upper_bound) of the confidence interval.
    """

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    standard_error = sample_std / np.sqrt(sample_size)

    degrees_of_freedom = sample_size - 1
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)

    margin_of_error = t_critical * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

data = [102, 105, 98, 108, 100, 103, 99, 101, 106, 104]
confidence_interval = calculate_confidence_interval(data)

print(f"Confidence Interval (95%): {confidence_interval}")

confidence_interval_99 = calculate_confidence_interval(data, 0.99)
print(f"Confidence Interval (99%): {confidence_interval_99}")

large_data = np.random.normal(100, 15, 1000)
large_confidence_interval = calculate_confidence_interval(large_data)

print(f"Confidence Interval (95%, Large Sample): {large_confidence_interval}")
```

# **Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
import numpy as np

def bayesian_inference(prior, likelihood, data):
    """
    Performs Bayesian inference using Bayes' Theorem.

    Args:
        prior (np.array): The prior probability distribution.
        likelihood (function): A function that calculates the likelihood of the data given a hypothesis.
        data (any): The observed data.

    Returns:
        np.array: The posterior probability distribution.
    """

    likelihood_values = np.array([likelihood(hypothesis, data) for hypothesis in range(len(prior))])

    unnormalized_posterior = prior * likelihood_values

    posterior = unnormalized_posterior / np.sum(unnormalized_posterior)

    return posterior

def coin_flip_likelihood(hypothesis, data):
    """
    Calculates the likelihood of observing data (heads/tails) given a coin bias.

    Args:
        hypothesis (int): The hypothesis (number of heads out of total flips, i.e. a bias).
        data (list): The observed data (e.g., [1, 0, 1] for heads, tails, heads).

    Returns:
        float: The likelihood.
    """
    heads = sum(data)
    tails = len(data) - heads
    p_heads = hypothesis / len(prior)

    return (p_heads ** heads) * ((1 - p_heads) ** tails)

prior = np.array([1/11] * 11)
data = [1, 1, 0, 1, 0, 1]

posterior = bayesian_inference(prior, coin_flip_likelihood, data)

for i, p in enumerate(posterior):
    print(f"Hypothesis (bias = {i/10:.1f}): Posterior Probability = {p:.4f}")

most_likely_hypothesis = np.argmax(posterior)
print(f"\nMost Likely Hypothesis (bias): {most_likely_hypothesis/10:.1f}")

prior2 = np.array([0.01, 0.01, 0.05, 0.08, 0.1, 0.15, 0.2, 0.15, 0.1, 0.08, 0.07])
posterior2 = bayesian_inference(prior2, coin_flip_likelihood, data)

print("\nResults with non-uniform prior:")
for i, p in enumerate(posterior2):
    print(f"Hypothesis (bias = {i/10:.1f}): Posterior Probability = {p:.4f}")

most_likely_hypothesis2 = np.argmax(posterior2)
print(f"\nMost Likely Hypothesis (bias) with non uniform prior: {most_likely_hypothesis2/10:.1f}")
```

# **Q12. Perform a Chi-square test for independence between two categorical variables in Python.**

```python
import numpy as np
from scipy.stats import chi2_contingency

def chi_square_independence_test(observed_data, alpha=0.05):
    """
    Performs a Chi-square test for independence between two categorical variables.

    Args:
        observed_data (list of lists or numpy.ndarray): The observed frequencies in a contingency table.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: (chi2_statistic, p_value, degrees_of_freedom, expected_frequencies, interpretation)
    """

    chi2_statistic, p_value, degrees_of_freedom, expected_frequencies = chi2_contingency(observed_data)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. There is significant evidence of an association between the two categorical variables."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. There is not enough evidence to conclude an association between the two categorical variables."

    return chi2_statistic, p_value, degrees_of_freedom, expected_frequencies, interpretation

observed_data = [
    [20, 30, 50],
    [10, 40, 60],
]

chi2_stat, p_val, df, expected, result_interpretation = chi_square_independence_test(observed_data)

print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Degrees of freedom: {df}")
print("Expected frequencies:\n", expected)
print(result_interpretation)

observed_data2 = [
    [100, 200],
    [150, 300],
]
chi2_stat2, p_val2, df2, expected2, result_interpretation2 = chi_square_independence_test(observed_data2)

print("\nSecond Example")
print(f"Chi-square statistic: {chi2_stat2:.4f}")
print(f"P-value: {p_val2:.4f}")
print(f"Degrees of freedom: {df2}")
print("Expected frequencies:\n", expected2)
print(result_interpretation2)
```

# **Q13. D Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.**

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculates the expected frequencies for a Chi-square test based on observed data.

    Args:
        observed_data (list of lists or numpy.ndarray): The observed frequencies in a contingency table.

    Returns:
        numpy.ndarray: The expected frequencies.
    """

    observed_data = np.array(observed_data)
    row_totals = observed_data.sum(axis=1)
    col_totals = observed_data.sum(axis=0)
    total = observed_data.sum()

    expected_frequencies = np.outer(row_totals, col_totals) / total
    return expected_frequencies

observed_data = [
    [20, 30, 50],
    [10, 40, 60],
]

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Data:\n", np.array(observed_data))
print("\nExpected Frequencies:\n", expected_frequencies)

observed_data2 = [
    [100, 200],
    [150, 300],
]

expected_frequencies2 = calculate_expected_frequencies(observed_data2)
print("\nSecond Example")
print("Observed Data:\n", np.array(observed_data2))
print("\nExpected Frequencies:\n", expected_frequencies2)
```

# **Q14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.**

```python
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed_frequencies, expected_frequencies=None, expected_probabilities=None, alpha=0.05):
    """
    Performs a goodness-of-fit test to compare observed data to an expected distribution.

    Args:
        observed_frequencies (list or numpy.ndarray): The observed frequencies.
        expected_frequencies (list or numpy.ndarray, optional): The expected frequencies.
        expected_probabilities (list or numpy.ndarray, optional): The expected probabilities.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: (chi2_statistic, p_value, interpretation)
    """

    observed_frequencies = np.array(observed_frequencies)

    if expected_frequencies is None:
        if expected_probabilities is None:
            raise ValueError("Either expected_frequencies or expected_probabilities must be provided.")
        else:
            expected_frequencies = np.array(expected_probabilities) * np.sum(observed_frequencies)

    chi2_statistic, p_value = chisquare(observed_frequencies, expected_frequencies)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. The observed data does not fit the expected distribution."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. The observed data fits the expected distribution."

    return chi2_statistic, p_value, interpretation

observed_frequencies = [12, 15, 8, 9, 11, 15]
expected_probabilities = [1/6] * 6

chi2_stat, p_val, result_interpretation = goodness_of_fit_test(observed_frequencies, expected_probabilities=expected_probabilities)

print("Example 1 (Fair Die Test):")
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

observed_frequencies2 = [25, 30, 45]
expected_frequencies2 = [20, 35, 45]

chi2_stat2, p_val2, result_interpretation2 = goodness_of_fit_test(observed_frequencies2, expected_frequencies=expected_frequencies2)

print("\nExample 2 (Specific Expected Frequencies):")
print(f"Chi-square statistic: {chi2_stat2:.4f}")
print(f"P-value: {p_val2:.4f}")
print(result_interpretation2)
```

# **Q16.  Implement an F-test using Python to compare the variances of two random samples.**

```python
import numpy as np
from scipy import stats

def f_test_variance(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two random samples.

    Args:
        sample1 (list or numpy.ndarray): The first sample data.
        sample2 (list or numpy.ndarray): The second sample data.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: (f_statistic, p_value, interpretation)
    """

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)

    if var1 > var2:
        f_statistic = var1 / var2
        dfn = n1 - 1
        dfd = n2 - 1
    else:
        f_statistic = var2 / var1
        dfn = n2 - 1
        dfd = n1 - 1

    p_value = 2 * min(stats.f.cdf(f_statistic, dfn, dfd), 1 - stats.f.cdf(f_statistic, dfn, dfd))

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. The variances are significantly different."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. There is not enough evidence to conclude that the variances are different."

    return f_statistic, p_value, interpretation

sample1 = [25, 30, 28, 35, 32, 29]
sample2 = [20, 26, 23, 28, 25, 22]

f_stat, p_val, result_interpretation = f_test_variance(sample1, sample2)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

sample3 = [10, 12, 11, 13, 15, 9]
sample4 = [20, 30, 25, 35, 40, 18]

f_stat2, p_val2, result_interpretation2 = f_test_variance(sample3, sample4)

print("\nSecond Example")
print(f"F-statistic: {f_stat2:.4f}")
print(f"P-value: {p_val2:.4f}")
print(result_interpretation2)
```

# **Q17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results. **

```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

def perform_anova(data, formula, alpha=0.05):
    """
    Performs an ANOVA test to compare means between multiple groups and interprets the results.

    Args:
        data (pandas.DataFrame): The data containing the dependent and independent variables.
        formula (str): The formula for the ANOVA test (e.g., 'dependent_var ~ independent_var').
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: (anova_table, interpretation)
    """

    model = ols(formula, data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    p_value = anova_table['PR(>F)'][0]
    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. There are significant differences between the means of the groups."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. There is not enough evidence to conclude differences between the means of the groups."

    return anova_table, interpretation

data = pd.DataFrame({
    'values': [25, 30, 28, 35, 32, 29, 20, 26, 23, 28, 25, 22, 38, 42, 40, 45, 43, 41],
    'groups': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C', 'C']
})

formula = 'values ~ groups'

anova_table, result_interpretation = perform_anova(data, formula)

print("ANOVA Table:\n", anova_table)
print("\nInterpretation:", result_interpretation)

np.random.seed(0)
group1 = np.random.normal(25, 5, 20)
group2 = np.random.normal(30, 5, 20)
group3 = np.random.normal(35, 5, 20)

data2 = pd.DataFrame({
    'values': np.concatenate([group1, group2, group3]),
    'groups': ['group1'] * 20 + ['group2'] * 20 + ['group3'] * 20
})

formula2 = 'values ~ groups'

anova_table2, result_interpretation2 = perform_anova(data2, formula2)

print("\nSimulated Data ANOVA Table:\n", anova_table2)
print("\nSimulated Data Interpretation:", result_interpretation2)
```

# **Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.**

```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, formula):
    """
    Checks the assumptions (normality, independence, equal variance) for ANOVA.

    Args:
        data (pandas.DataFrame): The data containing the dependent and independent variables.
        formula (str): The formula for the ANOVA test (e.g., 'dependent_var ~ independent_var').

    Returns:
        dict: A dictionary containing the results of the assumption checks.
    """

    model = ols(formula, data=data).fit()
    residuals = model.resid

    shapiro_stat, shapiro_p = stats.shapiro(residuals)
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    stats.probplot(residuals, plot=plt)
    plt.title('Q-Q Plot of Residuals')
    plt.subplot(1, 2, 2)
    sns.histplot(residuals, kde=True)
    plt.title('Histogram of Residuals')
    plt.show()

    group_col = formula.split('~')[1].strip()
    groups = [data[data[group_col] == group][formula.split('~')[0].strip()] for group in data[group_col].unique()]
    levene_stat, levene_p = stats.levene(*groups)

    independence_check = "Independence of observations should be ensured during study design."

    results = {
        "normality": {
            "shapiro_statistic": shapiro_stat,
            "shapiro_p_value": shapiro_p,
            "interpretation": "Residuals are normally distributed" if shapiro_p > 0.05 else "Residuals are not normally distributed"
        },
        "equal_variance": {
            "levene_statistic": levene_stat,
            "levene_p_value": levene_p,
            "interpretation": "Variances are equal" if levene_p > 0.05 else "Variances are not equal"
        },
        "independence": independence_check
    }

    return results

data = pd.DataFrame({
    'values': [25, 30, 28, 35, 32, 29, 20, 26, 23, 28, 25, 22, 38, 42, 40, 45, 43, 41],
    'groups': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C', 'C']
})

formula = 'values ~ groups'

assumption_results = check_anova_assumptions(data, formula)

for assumption, result in assumption_results.items():
    if isinstance(result, dict):
        print(f"\n{assumption.capitalize()} Check:")
        for key, value in result.items():
            print(f"  {key}: {value}")
    else:
        print(f"\n{assumption.capitalize()} Check:")
        print(f"  {result}")
```

# **Q20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

def perform_two_way_anova(data, formula):
    """
    Performs a two-way ANOVA test and visualizes the results.

    Args:
        data (pandas.DataFrame): The data containing the dependent and independent variables.
        formula (str): The formula for the two-way ANOVA test (e.g., 'dependent_var ~ factor1 * factor2').

    Returns:
        pandas.DataFrame: The ANOVA table.
    """

    model = ols(formula, data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    return anova_table

def visualize_two_way_anova(data, factor1_col, factor2_col, value_col):
    """
    Visualizes the results of a two-way ANOVA test.

    Args:
        data (pandas.DataFrame): The data containing the factors and values.
        factor1_col (str): The name of the column containing the first factor.
        factor2_col (str): The name of the column containing the second factor.
        value_col (str): The name of the column containing the values.
    """

    plt.figure(figsize=(12, 6))

    sns.pointplot(x=factor1_col, y=value_col, hue=factor2_col, data=data, capsize=.1)
    plt.title('Interaction Plot')
    plt.xlabel(factor1_col)
    plt.ylabel(value_col)
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.boxplot(x=factor1_col, y=value_col, hue=factor2_col, data=data)
    plt.title('Boxplot with Hue')
    plt.xlabel(factor1_col)
    plt.ylabel(value_col)
    plt.grid(True)
    plt.show()

data = pd.DataFrame({
    'values': [25, 30, 28, 35, 32, 29, 20, 26, 23, 28, 25, 22, 38, 42, 40, 45, 43, 41,
               30, 32, 33, 36, 34, 31, 22, 24, 26, 29, 27, 25, 40, 43, 41, 46, 44, 42],
    'factor1': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C', 'C',
                'A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C', 'C'],
    'factor2': ['X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X', 'X',
                'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y', 'Y']
})

formula = 'values ~ factor1 * factor2'

anova_table = perform_two_way_anova(data, formula)

print("Two-Way ANOVA Table:\n", anova_table)

visualize_two_way_anova(data, 'factor1', 'factor2', 'values')

np.random.seed(0)
factor1_values = ['A', 'B', 'C']
factor2_values = ['X', 'Y']
data_random = pd.DataFrame({
    'factor1': np.random.choice(factor1_values, 100),
    'factor2': np.random.choice(factor2_values, 100),
    'values': np.random.normal(25, 5, 100)
})

anova_table_random = perform_two_way_anova(data_random, formula)
print("\nRandom Data Two-Way ANOVA Table:\n", anova_table_random)
visualize_two_way_anova(data_random, 'factor1', 'factor2', 'values')
```

# **Q21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def visualize_f_distribution(dfn_list, dfd_list):
    """
    Visualizes the F-distribution for different degrees of freedom.

    Args:
        dfn_list (list): List of numerator degrees of freedom.
        dfd_list (list): List of denominator degrees of freedom.
    """

    x = np.linspace(0, 5, 500)

    plt.figure(figsize=(10, 6))

    for dfn, dfd in zip(dfn_list, dfd_list):
        y = f.pdf(x, dfn, dfd)
        plt.plot(x, y, label=f'dfn={dfn}, dfd={dfd}')

    plt.title('F-Distribution')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

dfn_list = [1, 5, 10]
dfd_list = [10, 20, 30]
visualize_f_distribution(dfn_list, dfd_list)

def f_test_example(dfn, dfd, f_value, alpha=0.05):
    """
    Demonstrates F-test usage and interpretation.

    Args:
        dfn (int): Numerator degrees of freedom.
        dfd (int): Denominator degrees of freedom.
        f_value (float): The calculated F-statistic.
        alpha (float, optional): Significance level. Defaults to 0.05.
    """

    p_value = 1 - f.cdf(f_value, dfn, dfd)

    print(f"\nF-test Example (dfn={dfn}, dfd={dfd}):")
    print(f"  F-statistic: {f_value:.4f}")
    print(f"  P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"  P-value < alpha ({alpha}). Reject null hypothesis.")
    else:
        print(f"  P-value >= alpha ({alpha}). Fail to reject null hypothesis.")

    x = np.linspace(0, max(f_value * 2, 5), 500)
    y = f.pdf(x, dfn, dfd)
    critical_value = f.ppf(1 - alpha, dfn, dfd)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'F-distribution (dfn={dfn}, dfd={dfd})')
    plt.axvline(critical_value, color='red', linestyle='--', label=f'Critical F: {critical_value:.2f}')
    plt.axvline(f_value, color='blue', linestyle='-', label=f'Observed F: {f_value:.2f}')
    plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.2)
    plt.title('F-test Visualization')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

f_test_example(dfn=5, dfd=20, f_value=3.5)
```

# **Q22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

def perform_one_way_anova_and_boxplot(data, formula):
    """
    Performs a one-way ANOVA test and visualizes the results with boxplots.

    Args:
        data (pandas.DataFrame): The data containing the dependent and independent variables.
        formula (str): The formula for the ANOVA test (e.g., 'dependent_var ~ independent_var').

    Returns:
        pandas.DataFrame: The ANOVA table.
    """

    model = ols(formula, data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    group_col = formula.split('~')[1].strip()
    value_col = formula.split('~')[0].strip()

    plt.figure(figsize=(10, 6))
    sns.boxplot(x=group_col, y=value_col, data=data)
    sns.swarmplot(x=group_col, y=value_col, data=data, color=".25")
    plt.title('One-Way ANOVA Results')
    plt.xlabel('Groups')
    plt.ylabel('Values')
    plt.grid(True)
    plt.show()

    return anova_table

data = pd.DataFrame({
    'values': [25, 30, 28, 35, 32, 29, 20, 26, 23, 28, 25, 22, 38, 42, 40, 45, 43, 41],
    'groups': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C', 'C']
})

formula = 'values ~ groups'

anova_table = perform_one_way_anova_and_boxplot(data, formula)

print("One-Way ANOVA Table:\n", anova_table)

np.random.seed(0)
group1 = np.random.normal(25, 5, 20)
group2 = np.random.normal(30, 5, 20)
group3 = np.random.normal(35, 5, 20)

data2 = pd.DataFrame({
    'values': np.concatenate([group1, group2, group3]),
    'groups': ['group1'] * 20 + ['group2'] * 20 + ['group3'] * 20
})

formula2 = 'values ~ groups'

anova_table2 = perform_one_way_anova_and_boxplot(data2, formula2)

print("\nRandom Data One-Way ANOVA Table:\n", anova_table2)
```

# **Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def simulate_and_test_means(mean1, std1, n1, mean2, std2, n2, alpha=0.05):
    """
    Simulates random data from two normal distributions and performs a t-test to compare the means.

    Args:
        mean1 (float): Mean of the first distribution.
        std1 (float): Standard deviation of the first distribution.
        n1 (int): Sample size of the first distribution.
        mean2 (float): Mean of the second distribution.
        std2 (float): Standard deviation of the second distribution.
        n2 (int): Sample size of the second distribution.
        alpha (float, optional): Significance level. Defaults to 0.05.
    """

    sample1 = np.random.normal(mean1, std1, n1)
    sample2 = np.random.normal(mean2, std2, n2)

    t_stat, p_value = stats.ttest_ind(sample1, sample2)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.hist(sample1, bins=20, alpha=0.6, label='Sample 1')
    plt.title('Sample 1 Distribution')
    plt.subplot(1, 2, 2)
    plt.hist(sample2, bins=20, alpha=0.6, label='Sample 2')
    plt.title('Sample 2 Distribution')
    plt.show()

    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"P-value < alpha ({alpha}). Reject the null hypothesis. There is a significant difference between the means.")
    else:
        print(f"P-value >= alpha ({alpha}). Fail to reject the null hypothesis. There is no significant difference between the means.")

mean1 = 50
std1 = 10
n1 = 100

mean2 = 55
std2 = 10
n2 = 100

simulate_and_test_means(mean1, std1, n1, mean2, std2, n2)

mean3 = 50
std3 = 10
n3 = 100

mean4 = 51
std4 = 10
n4 = 100

simulate_and_test_means(mean3, std3, n3, mean4, std4, n4)
```

# **Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.**

```python
import numpy as np
import scipy.stats as stats

def chi_square_variance_test(sample_variance, sample_size, hypothesized_variance, alpha=0.05):
    """
    Performs a Chi-square test for population variance and interprets the results.

    Args:
        sample_variance (float): The sample variance.
        sample_size (int): The sample size.
        hypothesized_variance (float): The hypothesized population variance.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: (chi2_statistic, p_value, interpretation)
    """

    degrees_of_freedom = sample_size - 1
    chi2_statistic = (degrees_of_freedom * sample_variance) / hypothesized_variance
    p_value = stats.chi2.cdf(chi2_statistic, degrees_of_freedom)

    if chi2_statistic < degrees_of_freedom:
      p_value = 2*p_value
    else:
      p_value = 2*(1-p_value)

    p_value = min(1, p_value)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. The population variance is significantly different from the hypothesized variance."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. There is not enough evidence to conclude that the population variance is different from the hypothesized variance."

    return chi2_statistic, p_value, interpretation

sample_data = [25, 30, 28, 35, 32, 29, 20, 26, 23, 28]
sample_variance = np.var(sample_data, ddof=1)
sample_size = len(sample_data)
hypothesized_variance = 20

chi2_stat, p_val, result_interpretation = chi_square_variance_test(sample_variance, sample_size, hypothesized_variance)

print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

hypothesized_variance2 = 50
chi2_stat2, p_val2, result_interpretation2 = chi_square_variance_test(sample_variance, sample_size, hypothesized_variance2)

print("\nSecond Example:")
print(f"Chi-square statistic: {chi2_stat2:.4f}")
print(f"P-value: {p_val2:.4f}")
print(result_interpretation2)
```

# **Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.**

```python
import numpy as np
import scipy.stats as stats

def z_test_proportions(count1, nobs1, count2, nobs2, alternative='two-sided', alpha=0.05):
    """
    Performs a Z-test for comparing proportions between two datasets or groups.

    Args:
        count1 (int): Number of successes in the first group.
        nobs1 (int): Total number of observations in the first group.
        count2 (int): Number of successes in the second group.
        nobs2 (int): Total number of observations in the second group.
        alternative (str, optional): 'two-sided', 'smaller', or 'larger'. Defaults to 'two-sided'.
        alpha (float, optional): Significance level. Defaults to 0.05.

    Returns:
        tuple: (z_statistic, p_value, interpretation)
    """

    prop1 = count1 / nobs1
    prop2 = count2 / nobs2
    prop_pooled = (count1 + count2) / (nobs1 + nobs2)
    se_pooled = np.sqrt(prop_pooled * (1 - prop_pooled) * (1/nobs1 + 1/nobs2))

    z_statistic = (prop1 - prop2) / se_pooled

    if alternative == 'two-sided':
        p_value = 2 * stats.norm.cdf(-np.abs(z_statistic))
    elif alternative == 'smaller':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'larger':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("alternative must be 'two-sided', 'smaller', or 'larger'")

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. There is a significant difference between the proportions."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. There is not enough evidence to conclude a difference between the proportions."

    return z_statistic, p_value, interpretation

count1 = 60
nobs1 = 100
count2 = 75
nobs2 = 120

z_stat, p_val, result_interpretation = z_test_proportions(count1, nobs1, count2, nobs2)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

z_stat_larger, p_val_larger, result_interpretation_larger = z_test_proportions(count1, nobs1, count2, nobs2, alternative='larger')
print("\nOne sided (larger) test:")
print(f"Z-statistic: {z_stat_larger:.4f}")
print(f"P-value: {p_val_larger:.4f}")
print(result_interpretation_larger)

z_stat_smaller, p_val_smaller, result_interpretation_smaller = z_test_proportions(count1, nobs1, count2, nobs2, alternative='smaller')
print("\nOne sided (smaller) test:")
print(f"Z-statistic: {z_stat_smaller:.4f}")
print(f"P-value: {p_val_smaller:.4f}")
print(result_interpretation_smaller)
```

# **Q26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def f_test_variance(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two datasets, interprets, and visualizes the results.

    Args:
        sample1 (list or numpy.ndarray): The first sample data.
        sample2 (list or numpy.ndarray): The second sample data.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: (f_statistic, p_value, interpretation)
    """

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)

    if var1 > var2:
        f_statistic = var1 / var2
        dfn = n1 - 1
        dfd = n2 - 1
    else:
        f_statistic = var2 / var1
        dfn = n2 - 1
        dfd = n1 - 1

    p_value = 2 * min(stats.f.cdf(f_statistic, dfn, dfd), 1 - stats.f.cdf(f_statistic, dfn, dfd))

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. The variances are significantly different."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. There is not enough evidence to conclude that the variances are different."

    visualize_f_test(sample1, sample2, f_statistic, dfn, dfd, alpha)

    return f_statistic, p_value, interpretation

def visualize_f_test(sample1, sample2, f_statistic, dfn, dfd, alpha):
    """
    Visualizes the F-test results.
    """

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(sample1, kde=True, label='Sample 1', color='skyblue')
    sns.histplot(sample2, kde=True, label='Sample 2', color='salmon')
    plt.title('Sample Distributions')
    plt.legend()

    plt.subplot(1, 2, 2)
    x = np.linspace(0, max(f_statistic * 2, 10), 500)
    pdf = stats.f.pdf(x, dfn, dfd)
    plt.plot(x, pdf, label=f'F-distribution (dfn={dfn}, dfd={dfd})')
    critical_value = stats.f.ppf(1 - alpha / 2, dfn, dfd)
    plt.axvline(critical_value, color='red', linestyle='--', label=f'Critical F: {critical_value:.2f}')
    plt.axvline(f_statistic, color='blue', linestyle='-', label=f'Observed F: {f_statistic:.2f}')
    plt.fill_between(x, pdf, where=(x >= critical_value), color='red', alpha=0.2)
    plt.title('F-test Visualization')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.legend()

    plt.tight_layout()
    plt.show()

sample1 = [25, 30, 28, 35, 32, 29]
sample2 = [20, 26, 23, 28, 25, 22]

f_stat, p_val, result_interpretation = f_test_variance(sample1, sample2)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

sample3 = [10, 12, 11, 13, 15, 9]
sample4 = [20, 30, 25, 35, 40, 18]

f_stat2, p_val2, result_interpretation2 = f_test_variance(sample3, sample4)

print("\nSecond Example:")
print(f"F-statistic: {f_stat2:.4f}")
print(f"P-value: {p_val2:.4f}")
print(result_interpretation2)
```

# **Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def chi_square_goodness_of_fit_simulation(expected_probabilities, sample_size, alpha=0.05):
    """
    Simulates data and performs a Chi-square goodness-of-fit test.

    Args:
        expected_probabilities (list or numpy.ndarray): The expected probabilities for each category.
        sample_size (int): The total sample size.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: (chi2_statistic, p_value, interpretation)
    """

    expected_probabilities = np.array(expected_probabilities)
    num_categories = len(expected_probabilities)

    simulated_data = np.random.choice(range(num_categories), size=sample_size, p=expected_probabilities)

    observed_frequencies = np.array([np.sum(simulated_data == i) for i in range(num_categories)])

    expected_frequencies = expected_probabilities * sample_size

    chi2_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.4f}) is less than alpha ({alpha}). Reject the null hypothesis. The simulated data does not fit the expected distribution."
    else:
        interpretation = f"The p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}). Fail to reject the null hypothesis. The simulated data fits the expected distribution."

    visualize_chi_square_goodness_of_fit(observed_frequencies, expected_frequencies)

    return chi2_statistic, p_value, interpretation

def visualize_chi_square_goodness_of_fit(observed_frequencies, expected_frequencies):
    """
    Visualizes the results of a Chi-square goodness-of-fit test.
    """
    categories = range(len(observed_frequencies))

    plt.figure(figsize=(10, 6))
    plt.bar(categories, observed_frequencies, label='Observed Frequencies', alpha=0.7)
    plt.plot(categories, expected_frequencies, marker='o', linestyle='-', color='red', label='Expected Frequencies')
    plt.xticks(categories)
    plt.xlabel('Categories')
    plt.ylabel('Frequencies')
    plt.title('Chi-square Goodness-of-Fit')
    plt.legend()
    plt.grid(True)
    plt.show()

expected_probabilities = [1/6] * 6
sample_size = 1000

chi2_stat, p_val, result_interpretation = chi_square_goodness_of_fit_simulation(expected_probabilities, sample_size)

print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result_interpretation)

expected_probabilities2 = [0.1, 0.2, 0.15, 0.25, 0.1, 0.2]
sample_size2 = 1000

chi2_stat2, p_val2, result_interpretation2 = chi_square_goodness_of_fit_simulation(expected_probabilities2, sample_size2)

print("\nSecond Example:")
print(f"Chi-square statistic: {chi2_stat2:.4f}")
print(f"P-value: {p_val2:.4f}")
print(result_interpretation2)
```
